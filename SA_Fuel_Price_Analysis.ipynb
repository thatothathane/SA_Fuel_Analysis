```python
import pandas as pd
import matplotlib.pyplot as plt
#Offical Inland 95 Unleaded Petrol Prices (ZAR per Litre)
#Below Shows the Journey of Price From 2020 to 2026
data = {
    'Date' : [
    '2020-01-01',
    '2021-01-01',
    '2022-01-05',
    '2022-07-06',
    '2023-01-04',
    '2024-01-03',
    '2025-01-01',
    '2026-02-04'
],
'Price_ZAR' : [
    16.16, 14.86, 19.61, 26.74, 21.40, 22.49, 21.59, 20.10
]
}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
print ("South African Petrol Price Dataset")
print (df)
#Calculate the percentage change between each date and it shows how prices hav changed overtime
#Formular (new - old) / old * 100
df['Percentage_Change'] = df['Price_ZAR'].pct_change()*100
print (df)
```

```python
import matplotlib.pyplot as plt
#The size of the graph is
plt.figure(figsize=(10,6))
#Plot Date vs Price
plt.plot(df['Date'], df['Price_ZAR'], marker='o', linestyle='-', color='blue')
#Add labels on
plt.title('South African Petrol Price Trend (2020-2026)')
plt.xlabel('Date')
plt.ylabel('Price in (R)')
plt.grid(True)
plt.show()
```

The graph shows fuel prices between 2020 to 2026, measured in Rands. The graph shows that petrol prices decreased in 2021, then increased and reached their highest point in 2023. After that the prices slowly went down again until 2026, but they are still higher than before.
